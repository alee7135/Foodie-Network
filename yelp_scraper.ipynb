```python
def scrape_single_page(bus_url):
    '''
    INPUT: restaurant url
    OUTPUT: list of people who reviewed the restaurant
    '''
    html = urlopen(bus_url).read()
    soup = BeautifulSoup(html, "lxml")
    reviewers = soup.select('a.user-display-name')
    reviewer_ids = set()
    for reviewer in reviewers:
        reviewer_ids.add(reviewer.get('href'))
    return reviewer_ids
```

```python
def scrape_get_successivepagelinks(bus_url):
    '''
    INPUT: restaurant url
    OUTPUT: list of restaurant page urls to be scraped one by one
    '''
    html = urlopen(bus_url).read()
    soup = BeautifulSoup(html, "lxml")
    review_count = re.findall(r'\d+', soup.find("span", attrs={'class':'review-count rating-qualifier'}).text)
    review_count = int(review_count[0])
    num_pages = review_count / 20
    
    prefix = bus_url + '?start='
    page_ids = []
    for page in xrange(num_pages + 1):  # need to include +1 extra page because python goes up to but not inclusive
        page_ids.append(prefix + str(page*20)) # only returns 20 results per page
    return page_ids
```

```python
def scrape_reviewers(bus_url):
    '''
    INPUT: restaurant url
    OUTPUT: list of all reviewers who reviewe this restaurant 
    '''
    pages = scrape_get_successivepagelinks(bus_url)
    time.sleep(30.0)
    all_reviewer_ids = set()
    for page in pages:
        all_reviewer_ids = all_reviewer_ids.union(scrape_single_page(page))
        time.sleep(30.0)
    return set(all_reviewer_ids)
```

```python
def page_link_generator(user_id, first_time=False, page_number=None):
    '''
    INPUT: reviewer id, whether it is the first page of reviewer, current page to scrape
    OUTPUT: actual link to reviewer's page(s) 
    '''
    # Can replace this later as a constant
    domain = 'http://www.yelp.com'
    # Pull out only user id
    userid = re.findall(r'userid=(.+$)', user_id)[0]
    # Create new link which pre-filters restaurants only 
    if first_time == True:
        link = domain + '/user_details_reviews_self?userid=' + userid + '&review_filter=category&category_filter=restaurants'
    else:
        link = domain + '/user_details_reviews_self?userid=' + userid + '&review_filter=category&category_filter=restaurants&rec_pagestart=%s' %(str(page_number))
    return link
```

```python
def scrape_reviewer_reviews(user_id, soup=None, page_number=None):
    '''
    INPUT: reviewer id
    OUTPUT: json object of all reviewer's reviews
        business_id, user_id, stars, text, date, rating (useful, funny, cool)
    '''

    if page_number==0:
        soup = soup
    else:
        link = page_link_generator(user_id, page_number=page_number)
        html = urlopen(link).read()
        soup = BeautifulSoup(html, "lxml")
        
    section = soup.find_all('ul', attrs={'class':'ytype ylist ylist-bordered reviews'})[0].find_all('li')
    check_in = 0  # need to re-initiate check-in because not all reviewers checked in for each restaurant
    for i in section:

        # business_id
        if i.find('a', attrs={'class':'biz-name'}):
            business_id = i.find('a', attrs={'class':'biz-name'})
            business_id = business_id.get('href')
            business_id = re.findall(r'^/biz/(.+$)', business_id)[0]

        # stars review
        if i.find('div', attrs={'class':"rating-very-large"}):
            stars = i.find('div', attrs={'class':"rating-very-large"})
            stars = stars.find('i').get('class')[1]
            stars = int(re.findall(r'\d+$', stars)[0])

        # texts 
        if i.find('p'):
            texts = i.find('p').text

        # date   
        if i.find('span', attrs={'class', 'rating-qualifier'}):
            date = i.find('span', attrs={'class', 'rating-qualifier'}).text
            date = re.findall(r"(\d+\/\d+\/\d+)", date)[0]

        # votes
        vote_types = {}
        if i.find_all('li', attrs={'class':'vote-item inline-block'}):
            emotion = ['useful', 'funny', 'cool']
            for vot in xrange(3): # useful, funny, cool
                vote = i.find_all('li', attrs={'class':'vote-item inline-block'})[vot]
                vote = vote.find('span', attrs={'class':'count'}).text
                if not vote:
                    vote_types[emotion[vot]] = 0
                else:
                    vote_types[emotion[vot]] = int(vote)

        # Check-In
        if i.find('div', attrs={'class':'review_tags'}):
            check_in = i.find('div', attrs={'class':'review_tags'})
            check_in = check_in.text
            check_in = int(re.findall(r'\d+', check_in)[0])


        check = i.find_all('a', attrs={'class':'biz-name'})
        if check:
            print business_id, stars, texts, date, vote_types, check_in, '\n'
            check_in = 0  # reinitiate check-in
    #         break
```

```python
def scrape_reviewer(user_id):
    '''
    INPUT: reviewer id
    OUTPUT: ***
    '''
    
    # This function needs to first check if User is already in database
 
    
    # Make request and get soup
    link = page_link_generator(user_id, first_time=True)
    html = urlopen(link).read()
    soup = BeautifulSoup(html, "lxml")
    # Retrieve number of pages 
    try:
        pages = soup.find('div', attrs={'class':'page-of-pages arrange_unit arrange_unit--fill'})
        pages = int(re.findall(r'([\d]+)\s+$', pages.text)[0])
    except:
        pages = 0
        print 'reviewer only has one page'
    
    # FIRST: Scrape reviewer stats
    scrape_reviewer_stats(user_id, soup)
    
    # Loop through each page and retrieve all restaurant ids which this particular reviewer reviewed
    for page in xrange(pages+1):
        scrape_reviewer_reviews(user_id, soup=soup, page_number=page)
        time.sleep(30.)
        
    return 
```
